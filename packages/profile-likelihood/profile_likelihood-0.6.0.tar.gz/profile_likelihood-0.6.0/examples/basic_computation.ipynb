# Basic computation

**We show a minimum implementation on how to compute the profile likelihood of the model.**

```python
import numpy as np
import matplotlib.pyplot as plt
from profile_likelihood import profile_likelihood

%matplotlib inline
```

## Define the model

The model is defined as the sum of trigonometric functions, in the form of
$$ f\left(t; \vec{\theta}\right) = \sin(\theta_0 t) + \cos(\theta_1 t), $$
where we sample at 3 time points $t = [1.0, 1.2, 2.0]^T$.
The default parameters are chosen to be $\vec{\theta} = [1.0, 2.5]^T$.
The data is set to be the predictions of the model at the default parameters and the standard deviation is chosen to be 30% of the data.
The implementation of this model is given in `model_trigonometric.py`.

```python
import model_trigonometric as model
import manifold
```

## Profile likelihood computation

```python
# Define the likelihood object
likelihood = profile_likelihood(model.residuals, model.nparams, model.npred)
```

As a default, the profile likelihood computation will be done to all the parameters in the model.
The fixed parameter values would be linearly spaced points in the range of $\theta \in (0, 10)$ with spacing $0.1$.

```python
# Compute the profile likelihood
results_all = likelihood.compute(model.best_fit)

# Print results
print(likelihood.results)
```

We can also specify which parameter to compute, e.g. $\theta_1$ only. The results would give an empty dictionary for $\theta_0$.

```python
results_1 = likelihood.compute(model.best_fit, idx=1)

# Print results
print(likelihood.results)
```
